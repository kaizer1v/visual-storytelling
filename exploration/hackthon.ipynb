```python
df = pd.read_csv("google_trends.csv", sep="|", index_col=0, parse_dates=True)
```

```python
# Convert everything to lower case
for i in range(2, 21):
    df['kw'+str(i)] = df['kw'+str(i)].str.lower()
```

```python
# We have grouped by year and month and calculated the total number of words that occur 

date_time = pd.to_datetime(df.index)
df.groupby([date_time.year, date_time.month]).describe()
```

```python
# Total number of keywords present in a month, for every year
df.T.groupby([date_time.year, date_time.month], axis=1).count()
```

```python
df['kw2'].groupby([date_time.year, date_time.month, df['kw2']]).count()
```
